- You can import data and read it. Here we use a .CSV (a specific excel like format) and save it into DataFrames called 'data'

- You can visualize all your data 

```
data
#The first row is called the header, it gives a unique name to each column
```

- You can also use the Pandas function [.head()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.head.html) the see only the first 5 lines 

```
#To use a library function we use a period after the name of the DataFrame 
data.head(5)
```

- You can display a specific column by calling it's name into quotes and store it in a variable called 'Genre'

```
data['Genre']
```

[.unique()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.unique.html)

```
data['Genre'].unique()
```

- You can find all the number values by using the Pandas function [.value_counts()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.value_counts.html) on that column

```
data['Genre'].value_counts()
```

```
data['Genre'].value_counts().plot(kind = 'bar')
```

```
data['Genre'].value_counts().plot(kind = 'pie')
```
